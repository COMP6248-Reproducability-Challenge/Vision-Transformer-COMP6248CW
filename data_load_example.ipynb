# data_load_example.ipynb
This file showed the format of loading datasets in this project.
Each part include a standard way to cut and resize images in datasets.

```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
%matplotlib inline
```

## Flowers102

```python
# https://pytorch.org/vision/stable/generated/torchvision.datasets.Flowers102.html
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),  # convert to tensor
    transforms.Resize(224),
    transforms.CenterCrop(224)
])
train_set = torchvision.datasets.Flowers102("./data", split= "train", download = True, transform=transform)
test_set = torchvision.datasets.Flowers102("./data", split= "test", download = True, transform=transform)
val_set = torchvision.datasets.Flowers102("./data", split= "val", download = True, transform=transform)
```

```python
print(train_set[0][0].shape)
plt.imshow(train_set[0][0].permute(1,2,0).numpy()[:,:,::])
plt.show()
```
